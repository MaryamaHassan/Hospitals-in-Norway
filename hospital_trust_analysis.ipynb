#### Employee Distribution and Regional Insights of Norwegian Hospital Trusts

```python

sns.countplot(x='Established Year', data=University_info, color='salmon',Label=None)
plt.title('Hospital Established Year')
plt.legend()
plt.tight_layout()
plt.show()
```

######  Year
###### The chart shows that the majority of these hospital trusts were established around 2001 and 2002. This timing aligns with the creation of regional health authorities, aimed at streamlining national healthcare functions as outlined by the National Institute of Public Health (FHI). The regional approach allowed for more efficient analysis of health data, service quality, and resource allocation at a broader level compared to focusing on individual cities, which would have been time-consuming. Consequently, this strategy contributed to a more organized and well-structured healthcare system across Norway.

```python
sns.displot(x='Employees',y='Regional Health Authority', data= University_info,hue='Employees')
```

###### Employees
###### The plot clearly shows that the Southern and Eastern Regional Health Authority has the highest number of employees. This is primarily because Oslo, the capital city of Norway, is situated in this region, leading to a larger population and higher demand for healthcare services. Additionally, Oslo is home to the country's largest hospital, Oslo University Hospital, where approximately 60% of Norway’s medical research is conducted, further increasing the concentration of healthcare professionals in the region.

```python
sns.barplot(x='Employees', data=University_info,palette='Accent')
plt.show()
```

###### The chart indicates that the total number of healthcare employees across Norway is estimated to be just over sixty thousand.
